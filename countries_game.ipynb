```python
from random import choice
from datetime import date


def load_countries():
    countries = []
    with open('countries.txt', mode = 'rt') as countries_file:
        for country in countries_file:
            country = country.strip()
            if len(country) >= 4 and len(country) <= 6:
                countries.append(country)
    return countries

def save_winners():
    todays_date = date.today()
    winner_text = (f'{todays_date} | {user_name} won in {ROUNDS}\n')
    with open('winners.txt', mode = 'at') as saved_winner:
        saved_winner.write(winner_text)
 
def print_winners():
    print('Here are the previous winners: ')
    with open('winners.txt', mode = 'rt') as saved_winner:
        for winner in saved_winner:
            print(winner, end = '')

countries = load_countries()
random_country = choice(countries)
hidden_country = ['_'] * len(random_country)

ROUNDS = 5

user_name = input('What is your name? ')
print_winners()


for round in range(1, ROUNDS + 1):

    random_letter = choice(random_country)
    random_letter_index = random_country.find(random_letter)
    hidden_country[random_letter_index] = random_letter
   
    
    print(hidden_country)
    
    guess = input("Can you guess the country? ")

    if guess == random_country:
        print("That's right, you win!")
        save_winners()
        break
        
    else:
        print(f"Try again! You've got {ROUNDS - round} rounds left.")
else:
    print(f'Better luck next time! The country was {random_country}.')

    
        
      
```
