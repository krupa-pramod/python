```python
import tkinter as tk
import random
root = tk.Tk()
root.title("krupa's gui")
root.geometry("500x500")

def play_game():
    num = random.randint(1,10)
    guess = int(entry.get())
    if num == guess:
        button.config(text = "You win! :)")
    else: 
        button.config(text = "You lost :(")

def play_again():
    play_game()
    label.config(text = "enter your guess and click the button again to see if you got it right this time...")

def hide_all():
    label.config(text = "bye!")
    label_welcome.pack_forget()
    label_guess.pack_forget()
    entry.pack_forget()
    button.pack_forget()
    button_yes.pack_forget()
    button_no.pack_forget()

label_welcome = tk.Label(root, text = "welcome!", bg = "#fadef6")
label_welcome.pack(pady=5)

label_guess = tk.Label(root, text = "guess a number from 1-10!", bg = "#fadef6")
label_guess.pack(pady=5)

entry = tk.Entry(root)
entry.pack(pady=5)

button = tk.Button(root, text="check to see if you won...", bg = "#cbdbf5", command=play_game)
button.pack(pady=5)

label = tk.Label(root, text = "want to play again?", bg = "#fadef6")
label.pack(pady=5)

button_yes = tk.Button(root, text="yes",  bg = "#cbdbf5", command=play_again)
button_yes.pack(pady=5)

button_no = tk.Button(root, text = "no",  bg = "#cbdbf5", command = hide_all)
button_no.pack(pady = 5)


root.mainloop()

```
